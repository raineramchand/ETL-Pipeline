1. What is the total revenue generated by VIP/Non-VIP guests?

```python
import sqlite3

# Connect to the database
conn = sqlite3.connect('hmart.db')

# Create a cursor
cursor = conn.cursor()

# Execute the SQL query
cursor.execute("""
    SELECT g.VIPStatus, SUM(f.Revenue) AS TotalRevenue
    FROM factTable f
    JOIN dimGuests g ON f.GuestID = g.GuestID
    GROUP BY g.VIPStatus;
""")


# Fetch all rows and print them
rows = cursor.fetchall()
for row in rows:
    print(row)

# Close the connection
conn.close()
```

2. Which country do most paying guests come from?

```python
import sqlite3

# Connect to the database
conn = sqlite3.connect('hmart.db')

# Create a cursor
cursor = conn.cursor()

# Execute the SQL query
cursor.execute("""
SELECT g.Country, COUNT(f.GuestID) AS NumberOfGuests, SUM(f.Revenue) AS TotalRevenue
FROM factTable f
JOIN dimGuests g ON f.GuestID = g.GuestID
GROUP BY g.Country
ORDER BY NumberOfGuests DESC
LIMIT 1;

""")


# Fetch all rows and print them
rows = cursor.fetchall()
for row in rows:
    print(row)

# Close the connection
conn.close()
```

4. What is the monthly revenue trend for the past year?

```python
import sqlite3

# Connect to the database
conn = sqlite3.connect('hmart.db')

# Create a cursor
cursor = conn.cursor()

# Execute the SQL query
cursor.execute("""
SELECT strftime('%Y-%m', p.PaymentDate) AS Month, SUM(f.Revenue) AS MonthlyRevenue
FROM factTable f
JOIN dimPayments p ON f.PaymentID = p.PaymentID
WHERE p.PaymentDate >= date('now', '-12 months')
GROUP BY strftime('%Y-%m', p.PaymentDate)
ORDER BY Month;

""")

# Fetch all rows and print them
rows = cursor.fetchall()
for row in rows:
    print(row)

# Close the connection
conn.close()
```

5. How much revenue is generated from guests with different stay durations?

```python
import sqlite3

# Connect to the database
conn = sqlite3.connect('hmart.db')

# Create a cursor
cursor = conn.cursor()

# Execute the SQL query
cursor.execute("""
SELECT 
    CASE 
        WHEN b.StayDuration BETWEEN 1 AND 3 THEN '1-3 days'
        WHEN b.StayDuration BETWEEN 4 AND 7 THEN '4-7 days'
        ELSE '8+ days'
    END AS StayDurationRange, 
    SUM(f.Revenue) AS TotalRevenue
FROM factTable f
JOIN dimBooking b ON f.BookingID = b.BookingID
GROUP BY StayDurationRange;


""")

# Fetch all rows and print them
rows = cursor.fetchall()
for row in rows:
    print(row)

# Close the connection
conn.close()
```
